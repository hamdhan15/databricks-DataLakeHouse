#Customer Dim Data

##Data Business Transformation

```python
query=("""
       SELECT
       ROW_NUMBER() OVER(ORDER BY cm.customer_id) as customer_key,
       cm.customer_id,
       cm.customer_name,
       cm.customer_marital_status,
       CASE 
       WHEN cm.customer_gender = 'n/a' THEN ca.gender
       ELSE cm.customer_gender
       END AS customer_gender,
       ca.birth_date,
       lo.country
       FROM
       workspace.silver.crm_cust_info cm
       left join workspace.silver.erp_cust_az12 ca
       on cm.customer_id=SUBSTRING(ca.customer_id,6)
       left join workspace.silver.erp_loc_az12 lo
       on cm.customer_id=(SUBSTRING(lo.customer_id,6))
       """)

df=spark.sql(query)
df.display()
```

##Loading Data into Gold Layer

```python
(
  df.write
  .format("delta")
  .mode("overwrite")
  .saveAsTable("workspace.gold.dim_gold_customer")
)
```
