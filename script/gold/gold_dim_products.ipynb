#Product Dim Data

##Business Transformation Logic

```python
query=("""
       SELECT
       ROW_NUMBER() OVER (ORDER BY pm.product_id) AS product_key,
       pm.product_id,
       pm.product_key as product_number,
       pm.product_category_id,
       pc.category as product_category,
       pc.sub_category as product_sub_category,
       pm.product_name,
       pm.product_cost,
       pm.product_line,
       pm.product_start_date,
       pm.product_end_date,
       pc.maintenance as product_maintenance
       FROM
       workspace.silver.crm_prd_info pm
       LEFT JOIN workspace.silver.erp_px_cat_az12 pc
       ON pm.product_category_id = pc.id
       """)
df=spark.sql(query)
# df.display()
```

```python
df=df.filter("product_category is not null")
# df.display()
```

##Loading Data into Gold Layer

```python
(
  df.write
  .mode("overwrite")
  .format("delta")
  .saveAsTable("workspace.gold.dim_gold_product")
)
```
