#Importing Sales_details Data

##Applying Business Transformation Logic

```python
query="""
SELECT
sd.sales_order_number,
cd.customer_key,
pd.product_key,
sd.sales_order_date,
sd.sales_ship_date,
sd.sales_due_date,
sd.total_sales,
sd.sales_quantity,
sd.sales_price
FROM
workspace.silver.crm_sales_details sd
LEFT JOIN gold.dim_gold_product pd
on sd.sales_product_key=pd.product_number
LEFT JOIN gold.dim_gold_customer cd
on sd.sales_customer_id=cd.customer_id
"""
df=spark.sql(query)
df.display()
```

## Writing data to Gold Layer

```python
(
  df.write
  .mode("overwrite")
  .format("delta")
  .saveAsTable("workspace.gold.dim_gold_sales")
)
```
